# An√°lisis del histograma

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

imagen_1 = cv2.imread("Histograma_1.png", cv2.IMREAD_GRAYSCALE)
plt.imshow(imagen_1, cmap='gray')
plt.show()
```

```python
imagen_2 = cv2.imread("Histograma_2.png", cv2.IMREAD_GRAYSCALE)
plt.imshow(imagen_2, cmap='gray')
plt.show()
```

```python
hist = cv2.calcHist([imagen_1],[0],None,[256],[0,256])
plt.plot(hist)
plt.title("Histograma de imagen 1")
plt.xlabel("Valor de gris")
plt.ylabel("Frecuencia de p√≠xeles")
plt.show()
```

```python
hist = cv2.calcHist([imagen_2],[0],None,[256],[0,256])
plt.plot(hist)
plt.title("Histograma de imagen 2")
plt.xlabel("Valor de gris")
plt.ylabel("Frecuencia de p√≠xeles")
plt.show()
```

### Explique las formas de cada uno de los histogramas 

* El histograma de la primera imagen muestra un pico marcado entre los valores 130 y 140, lo que indica que predominan los tonos grises medios y que no hay gran variaci√≥n en los niveles de gris entre los p√≠xeles. Por otro lado, el histograma de la segunda imagen presenta dos concentraciones principales, un pico pronunciado alrededor de 55 y una distribuci√≥n m√°s amplia aproximadamente entre 100 hasta 180 aproximadamente. Esto sugiere dos grupos principales de intensidades en la imagen, asociados a las diferentes fibras presentes.

### Observando √∫nicamente el histograma de muscle.png (Imagen 2), ¬øc√≥mo se puede determinar cu√°les fibras musculares son m√°s densas (ocupan una superficie m√°s grande en la imagen) entre las rojas (oscuras) y las blancas (grises)? Y ¬øcu√°les fibras tienen un color m√°s uniforme (m√°s homog√©neo)?

* Las fibras musculares rojas (oscuras) son m√°s homog√©neas que las fibras blancas (grises). El pico de concentracion alrededor de 50 y 55 muestra que las fibras oscuras tienen intensidades muy similares con poca variaci√≥n. En cambio, las fibras blancas muestran una monta√±a peque√±a entre 150 y 250, se√±alando gran variabilidad en sus intensidades. 

# Operaciones aritm√©ticas


```python
imagen_1 = cv2.imread("Operaciones_1.png", cv2.IMREAD_GRAYSCALE)
plt.imshow(imagen_1, cmap='gray')
plt.show()
```

```python
hist = cv2.calcHist([imagen_1],[0],None,[256],[0,256])
plt.plot(hist)
plt.title("Histograma de imagen 1")
plt.xlabel("Valor de gris")
plt.ylabel("Frecuencia de p√≠xeles")
plt.show()
```

```python
hist = cv2.calcHist([imagen_2],[0],None,[256],[0,256])
plt.plot(hist)
plt.title("Histograma de imagen 2")
plt.xlabel("Valor de gris")
plt.ylabel("Frecuencia de p√≠xeles")
plt.show()
```

### Explique las formas de cada uno de estos histogramas.

* En el histograma de la imagen 1 se observa que hay dos picos pronunciados: uno alrededor de 50 y otro alrededor de 170. Adem√°s, presenta una distribuci√≥n baja entre ambos picos, por lo que se puede concluir que la imagen tiene dos grupos principales de intensidades. Por otro lado, el histograma de la imagen 2 muestra √∫nicamente dos picos grandes en 0 y 255, lo que indica que la imagen est√° compuesta pr√°cticamente por negro y blanco.

```python
Adiccion = cv2.add(imagen_1, imagen_2)
plt.imshow(Adiccion, cmap='gray')
plt.title('Adiccion de imagen 2 en imagen 1')
plt.show()
```

```python
hist = cv2.calcHist([Adiccion],[0],None,[256],[0,256])
plt.plot(hist)
plt.title("Histograma de adiccion")
plt.xlabel("Valor de gris")
plt.ylabel("Frecuencia de p√≠xeles")
plt.show()
```

### Justifique los valores de los niveles de gris de esta imagen.

* Al hacer la adici√≥n de la imagen 2 a la imagen 1, podemos observar que el histograma de la imagen 1 se mantiene en gran parte, ya que siguen apareciendo los dos picos principales y la distribuci√≥n entre los picos. Por otro lado, los p√≠xeles blancos aumentan significativamente debido a la contribuci√≥n de la imagen 2, lo que genera un pico mucho mayor cerca de 255. Como consecuencia de esto, en la gr√°fica el resto de los picos y la distribuci√≥n del histograma de la imagen 1 se ven relativamente m√°s bajos.

```python
Sustraccion_1 = cv2.subtract(imagen_1, imagen_2)
plt.imshow(Sustraccion_1, cmap='gray')
plt.title('Sustraccion de imagen 1 - imagen 2')
plt.show()
```

```python
Sustraccion_2 = cv2.subtract(imagen_2, imagen_1)
plt.imshow(Sustraccion_2, cmap='gray')
plt.title('Sustraccion de imagen 2 - imagen 1')
plt.show()
```

```python
hist = cv2.calcHist([Sustraccion_1],[0],None,[256],[0,256])
plt.plot(hist)
plt.title("Histograma de Sustraccion 1")
plt.xlabel("Valor de gris")
plt.ylabel("Frecuencia de p√≠xeles")
plt.show()
```

```python
hist = cv2.calcHist([Sustraccion_2],[0],None,[256],[0,256])
plt.plot(hist)
plt.title("Histograma de Sustraccion 2")
plt.xlabel("Valor de gris")
plt.ylabel("Frecuencia de p√≠xeles")
plt.show()
```

### ¬øC√≥mo afecta al resultado la inversi√≥n de los operandos?

* Al invertir los operandos cambia totalmente el resultado, tanto de las im√°genes como de los histogramas. Esto ocurre principalmente porque la resta no es una operaci√≥n conmutativa, por lo que intercambiar los operandos determina cu√°l imagen le quita intensidad a la otra. Por ejemplo, en la sustracci√≥n 1 podemos ver que la resta afecta principalmente la zona donde la imagen 2 es blanca, haciendo que esa regi√≥n en la imagen 1 se oscurezca hasta negro y aumentando las ocurrencias cerca de 0 en el histograma.

### ¬øEs posible prever el histograma de la imagen que se obtendr√° con una operaci√≥n aritm√©tica sobre estas im√°genes operandos? ¬øPor qu√©?

* No es posible prever el histograma de la imagen resultante √∫nicamente a partir de los histogramas de las im√°genes operando. Esto se debe a que el histograma solo contiene informaci√≥n sobre el n√∫mero de ocurrencias en cada intensidad, pero no incluye informaci√≥n espacial ni la correlaci√≥n entre los p√≠xeles.

# Negativo

```python
imagen_1 = cv2.imread("Negativo.png", cv2.IMREAD_GRAYSCALE)
plt.imshow(imagen_1, cmap='gray')
plt.show()
```

```python
hist = cv2.calcHist([imagen_1],[0],None,[256],[0,256])
plt.plot(hist)
plt.title("Histograma de imagen 1")
plt.xlabel("Valor de gris")
plt.ylabel("Frecuencia de p√≠xeles")
plt.show()
```

### ¬øC√≥mo es la din√°mica de la imagen?

* Es posible observar c√≥mo la imagen tiene una din√°mica concentrada principalmente en intensidades bajas y medias, con un pico pronunciado alrededor de 100 y un pico menor cerca de 30. Esto nos indica que la imagen est√° dominada por tonos oscuros y grises medios, adem√°s de que no aprovecha completamente todo el rango de valores.

```python
imagen_1_negativo = cv2.bitwise_not(imagen_1)
plt.imshow(imagen_1_negativo, cmap='gray')
plt.show()
```

```python
hist = cv2.calcHist([imagen_1_negativo],[0],None,[256],[0,256])
plt.plot(hist)
plt.title("Histograma de imagen 1 negativa")
plt.xlabel("Valor de gris")
plt.ylabel("Frecuencia de p√≠xeles")
plt.show()
```

### ¬øC√≥mo es este histograma con respecto al de la imagen original?

* Observando el histograma del negativo, podemos notar que resulta ser el espejo del histograma de la imagen original. Esto se debe a que se convirtieron los p√≠xeles que eran negros en blancos y viceversa.

```python
Rmin = np.min(imagen_1)
Rmax = np.max(imagen_1)
imagen_1_expansion = ((imagen_1 - Rmin) / (Rmax - Rmin) * 255).astype(np.uint8)

plt.imshow(imagen_1_expansion, cmap='gray')
plt.show()
```

```python
hist = cv2.calcHist([imagen_1_expansion],[0],None,[256],[0,256])
plt.plot(hist)
plt.title("Histograma expandido de imagen 1")
plt.xlabel("Valor de gris")
plt.ylabel("Frecuencia de p√≠xeles")
plt.show()
```

```python
Rmin = np.min(imagen_1_negativo)
Rmax = np.max(imagen_1_negativo)
imagen_1_expansion_negativo = ((imagen_1_negativo - Rmin) / (Rmax - Rmin) * 255).astype(np.uint8)

plt.imshow(imagen_1_expansion_negativo, cmap='gray')
plt.show()
```

```python
hist = cv2.calcHist([imagen_1_expansion_negativo],[0],None,[256],[0,256])
plt.plot(hist)
plt.title("Histograma expandido de imagen 1 negativa")
plt.xlabel("Valor de gris")
plt.ylabel("Frecuencia de p√≠xeles")
plt.show()
```

### Explique el resultado

* Aplicando la expansi√≥n de los histogramas de ambas im√°genes, podemos observar que los niveles de gris que antes se concentraban en un rango muy reducido ahora se distribuyen a lo largo de casi todo el intervalo. Por eso, el histograma se estira horizontalmente, lo que permite ver c√≥mo las zonas m√°s oscuras se vuelven m√°s oscuras y las zonas claras m√°s claras.




# Cuantificaci√≥n

```python
imagen_1 = cv2.imread("Cuantificacion.png", cv2.IMREAD_GRAYSCALE)
plt.imshow(imagen_1, cmap='gray')
plt.show()
```

```python
def bit_plane_slicing(imagen, k):
    plane = np.full((imagen.shape[0], imagen.shape[1]), 2 ** k, np.uint8)
    return cv2.bitwise_and(plane, imagen)

for i in range(0, 8):
    plano = bit_plane_slicing(imagen_1, i)
    plt.figure()
    plt.imshow(plano, cmap="gray")
    plt.title(f"Plano binario {i}")
    plt.show()
```

### Identifique el plano m√°s significativo y el menos significativo. ¬øCu√°l es su conclusi√≥n?

* Al visualizar todos los planos binarios, podemos concluir que el plano m√°s significativo es el plano 7, porque conserva la mayor parte de la estructura y las formas principales, mientras que el menos significativo es el plano 0, ya que es el que presenta m√°s ruido y no aporta mucha informaci√≥n. Adem√°s, es posible notar que, a medida que se desciende entre planos, se va perdiendo informaci√≥n y comienza a predominar el ruido.

```python
def image_quantization(imagen, k):
    plane = np.full((imagen.shape[0], imagen.shape[1]), 255-2 ** k, np.uint8)
    return cv2.bitwise_and(plane, imagen)

Cuantificacion_1 = image_quantization(imagen_1, 1)
plt.imshow(Cuantificacion_1, cmap="gray")
plt.title("Cuantificacion 1")
plt.show()

Cuantificacion_3 = image_quantization(imagen_1, 3)
plt.imshow(Cuantificacion_3, cmap="gray")
plt.title("Cuantificacion 3")
plt.show()

Cuantificacion_5 = image_quantization(imagen_1, 5)
plt.imshow(Cuantificacion_5, cmap="gray")
plt.title("Cuantificacion 5")
plt.show()
```

### Visualice las im√°genes resultado y comente sus conclusiones.

* Al aplicar la cuantificaci√≥n con $ k = 1 $, la imagen conserva bien su estructura general y los detalles no se pierden de forma notable. Con 
$ k = 3 $, ya se puede empezar a notar c√≥mo la imagen comienza a perder detalle y aparecen contornos falsos. Por √∫ltimo, con $ k = 5 $ se observa una imagen considerablemente m√°s pixelada y con a√∫n m√°s contornos falsos que en el caso anterior.
