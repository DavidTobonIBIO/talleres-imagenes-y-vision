# An√°lisis del histograma

```python
import matplotlib.pyplot as plt
import numpy as np

imagen_1 = plt.imread("Histograma_1.png")
plt.imshow(imagen_1)
plt.show()
```

```python
imagen_2 = plt.imread("Histograma_2.png")
plt.imshow(imagen_2)
plt.show()
```

```python
histograma_1, bin_edges_1 = np.histogram(imagen_1, bins=256, range=(0, 1))

plt.plot(bin_edges_1[0:-1], histograma_1)
plt.title("Histograma de imagen 1")
plt.xlabel("Valor de gris")
plt.ylabel("Frecuencia de p√≠xeles")
plt.show()
```

```python
histograma_2, bin_edges_2 = np.histogram(imagen_2, bins=256, range=(0,1))

plt.plot(bin_edges_2[0:-1], histograma_2)
plt.title("Histograma de imagen 2")
plt.xlabel("Valor de gris")
plt.ylabel("Frecuencia de p√≠xeles")
plt.show()
```

### Explique las formas de cada uno de los histogramas 

* El histograma de la primera imagen muestra una concentraci√≥n alrededor del valor 0.5, alcanzando aproximadamente 12000 p√≠xeles de frecuencia. Esta distribuci√≥n con un pico alto y estrecho es caracter√≠stica de una imagen con predominancia de tonos grises medios uniformes. Por otro lado, el histograma de la segunda imagen presenta 2 diferentes concentraciones. La primera es un pico en 0.2 con aproximadamente 3500 p√≠xeles (fibras rojas oscuras), y la segunda es una meseta entre 0.6 y 0.8 con frecuencias de 1500 a 1700 p√≠xeles (fibras blancas grises). Esta forma indica dos poblaciones distintas de tejido muscular con diferentes intensidades.
